```python
import matplotlib.pyplot as plt
import numpy as np


TRAINING_LOSS_DIR = "/scratch/gpfs/sbfisher/EM300/results/"
TEST_LOSS_DIR = "/scratch/gpfs/sbfisher/EM300/inverse_loss/"
```

```python
train_loss_fn = TRAINING_LOSS_DIR + "EMInverseTrainLoss.npy"
val_loss_fn = TRAINING_LOSS_DIR + "EMInverseValidationLoss.npy"

train_loss = np.load(train_loss_fn)
val_loss = np.load(val_loss_fn)

print(train_loss.shape, val_loss.shape)

num_epochs = 40
epochs = [i for i in range(num_epochs)]
# plot the loss vs epoch, L_b and L_s

plt.plot(epochs, train_loss[:,0])
plt.plot(epochs, val_loss[:,0], linestyle="--")
plt.title("Ls vs. epoch, train and val")
plt.legend(["Train Ls", "Val Ls"])
plt.xlabel("epoch")
plt.ylabel("Ls")

plt.annotate("Weight of Lb->1", xy=(10, 0.038), xytext=(11, 0.1),
         bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='none'),
         # arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.5", color="black"))
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.5", color="black"))
# plt.annotate("Weight of Lb->10", xy=(20, 0.04), xytext=(23, 0.15),
#          bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='none'),
#          # arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.5", color="black"))
#          arrowprops=dict(arrowstyle="->", color="black"))
# plt.annotate("Weight of Lb->40", xy=(30, 0.048), xytext=(32, 0.1),
#          bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='none'),
#          # arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.5", color="black"))
#          arrowprops=dict(arrowstyle="->", color="black"))
plt.savefig(TRAINING_LOSS_DIR + "Ls_vs_epoch_initial.png", dpi=300)
plt.figure()

plt.plot(epochs, train_loss[:,1])
plt.plot(epochs, val_loss[:,1], linestyle="--")
plt.title("Lb vs. epoch, train and val")
plt.legend(["Train Lb", "Val Lb"])
plt.annotate("Weight of Lb->1", xy=(10, 0.0225), xytext=(11, 0.02),
         bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='none'),
         # arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.5", color="black"))
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.5", color="black"))
# plt.annotate("Weight of Lb->10", xy=(20, 0.0038), xytext=(21, 0.006),
#          bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='none'),
#          # arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.5", color="black"))
#          arrowprops=dict(arrowstyle="->", color="black"))
# plt.annotate("Weight of Lb->40", xy=(29, 0.0009), xytext=(31, 0.003),
#          bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='none'),
#          # arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.5", color="black"))
#          arrowprops=dict(arrowstyle="->", color="black"))
plt.xlabel("epoch")
plt.ylabel("Lb")
plt.savefig(TRAINING_LOSS_DIR + "Lb_vs_epoch_initial.png", dpi=300)
```

```python
# Test loss for threshold = 0.9 is 0.09574824834334611
inverse_val_loss = np.load(TEST_LOSS_DIR + "inverse_val_losses.npy")
thresholds = [round(0.1 + x * 0.05, 2) for x in range(int((0.9 - 0.1) / 0.05) + 1)]

print(inverse_val_loss.shape, len(thresholds))

plt.plot(thresholds, inverse_val_loss)
plt.xlabel("Threshold Cutoff Value")
plt.ylabel("Validation dataset loss")
plt.title("Validation loss vs. Threshold Value")
plt.savefig(TRAINING_LOSS_DIR + "threshold_vs_val_loss_initial.png", dpi=300)
```
